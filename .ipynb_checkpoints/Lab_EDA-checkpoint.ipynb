```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
df
```

<h1>Part 1</h1>

```python
# Frequency table for 'MSZoning'
frequency_table = df['category'].value_counts()

# Calculating the proportion of each unique value in the 'MSZoning'
proportion_table = df['category'].value_counts(normalize=True)

frequency_table
```

```python
proportion_table
```

```python
# Plotting a bar chart using the values from the frequency table, with colors sourced from the "Set3" palette
sns.barplot(x=frequency_table.index, y=frequency_table.values, palette="Set3");
```

```python
sns.countplot(data=df, x='category', palette="Set3");
```

```python
df_category = df.groupby("category").size().reset_index(name="products").sort_values("products", ascending=False).head(20)
df_category                                                                                                             
```

```python
df_top = df.groupby("category").size().reset_index(name="products").sort_values("products", ascending=False).head(5)
```

```python
sns.barplot(x=df_top["category"], y=df_top["products"], palette="Set3");
```

```python
'''The Sports & Outdoor category dominates with regards to products listed per category'''

# Plotting a pie chart of the 'category' column value counts, with percentage labels, 
df_category["products"].plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
```

<h1>Part 2</h1>

```python
df["price"].describe().round(2)
```

```python
df["price"].var()
```

```python
df["price"].std()
```

```python
'''There is evidence of extreme outliers to the top end of the price, as the min and max values are very far apart and 
75% of the values are under 45.99. Also mean and median are very far apart form each other'''
```

```python
quantiles_price = df['price'].quantile([0.25, 0.5, 0.75])
quantiles_price
```

```python
# Discretizing 'SalePrice' into 4 categories
bins = [0, 9.99, 19.09, 45.99, df['price'].max()]
df['price_category'] = pd.cut(df['price'], bins=bins, include_lowest=True)
```

```python
df.price_category.value_counts()
```

```python
sns.histplot(df['price'], kde=True, bins=4, color="salmon")
```
